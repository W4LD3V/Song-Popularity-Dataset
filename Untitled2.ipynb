```python
target = 'song_popularity'
features = df.drop(columns=[target, 'song_name']).columns

X = df[features]
y = df[target]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# z = x - μ / σ
# z = data point - mean / standart deviation
z_scores = np.abs(stats.zscore(X_scaled))

threshold = 3

outliers = (z_scores > threshold)

# np.any(outliers, axis=1) checks each row of the outliers array.
# It returns True if any value in that row is True (indicating that the row contains at least one outlier),
# and False otherwise.
# ~ is the bitwise NOT operator, which inverts the boolean values.
# So, True becomes False and vice versa.
# Therefore, mask will be True for rows without outliers and False for rows with at least one outlier.
mask = ~np.any(outliers, axis=1)

X_cleaned = X[mask]
y_cleaned = y[mask]

print("Original DataFrame shape:", X.shape)
print("Cleaned DataFrame shape:", X_cleaned.shape)

X_train, X_test, y_train, y_test = train_test_split(X_cleaned, y_cleaned, test_size=0.2, random_state=42)

print("Before Scaling (first 5 rows):")
print(tabulate(X_train.head(), headers='keys', tablefmt='pretty'))

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nAfter Scaling (first 5 rows):")
print(tabulate(pd.DataFrame(X_train_scaled, columns=features).head(), headers='keys', tablefmt='pretty'))

print("Number of outliers in each feature:")
print(np.sum(outliers, axis=0))
```

```python
df = df.drop(columns=['song_name'])
correlation_matrix = df.corr()

plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
model = LinearRegression()
model.fit(X_train_scaled, y_train)

y_pred_train = model.predict(X_train_scaled)
y_pred_test = model.predict(X_test_scaled)

train_mse = mean_squared_error(y_train, y_pred_train)
test_mse = mean_squared_error(y_test, y_pred_test)
train_r2 = r2_score(y_train, y_pred_train)
test_r2 = r2_score(y_test, y_pred_test)

print(f'\nTrain MSE: {train_mse}')
print(f'Test MSE: {test_mse}')
print(f'Train R^2: {train_r2}')
print(f'Test R^2: {test_r2}')

plt.figure(figsize=(14, 7))

# Training
plt.subplot(1, 2, 1)
plt.scatter(y_train, y_pred_train, alpha=0.5, label='Train Data', color='blue')
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], color='red', linestyle='--', label='Ideal Fit')
plt.xlabel('Actual Song Popularity')
plt.ylabel('Predicted Song Popularity')
plt.title('Training Set: Actual vs Predicted')
plt.legend()

# Testing
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pred_test, alpha=0.5, label='Test Data', color='green')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Ideal Fit')
plt.xlabel('Actual Song Popularity')
plt.ylabel('Predicted Song Popularity')
plt.title('Testing Set: Actual vs Predicted')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
models = {
    'Random Forest Regressor': RandomForestRegressor(n_estimators=100, random_state=42),
    'AdaBoost Regressor': AdaBoostRegressor(n_estimators=50, random_state=42),
    'Gradient Boosting Regressor': GradientBoostingRegressor(n_estimators=50, random_state=42),
    'XGBoost Regressor': xgb.XGBRegressor(n_estimators=50, random_state=42),
    'LightGBM Regressor': lgb.LGBMRegressor(n_estimators=50, random_state=42, force_col_wise=True),
    'CatBoost Regressor': cb.CatBoostRegressor(n_estimators=50, random_state=42, verbose=0)
}

def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, rmse, mae, r2

for model_name, model in models.items():
    mse, rmse, mae, r2 = evaluate_model(model, X_train_scaled, y_train, X_test_scaled, y_test)
    print(f'{model_name} - MSE: {mse}, RMSE: {rmse}, MAE: {mae}, R²: {r2}')

param_grid_rf = {
    'n_estimators': [50, 100],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5]
}

grid_search_rf = GridSearchCV(RandomForestRegressor(random_state=42), param_grid_rf, cv=3, n_jobs=-1, verbose=0)
grid_search_rf.fit(X_train_scaled, y_train)

best_rf = grid_search_rf.best_estimator_
mse, rmse, mae, r2 = evaluate_model(best_rf, X_train_scaled, y_train, X_test_scaled, y_test)
print(f'\nBest Random Forest Regressor - MSE: {mse}, RMSE: {rmse}, MAE: {mae}, R²: {r2}')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense


target = 'song_popularity'
features = df.drop(columns=[target, 'song_name']).columns

X = df[features].values
y = df[target].values

print("Features shape:", X.shape)
print("Target shape:", y.shape)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled = X_scaled.reshape((X_scaled.shape[0], 1, X_scaled.shape[1]))

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=1)

y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

test_mse = mean_squared_error(y_test, y_pred_test)
test_r2 = r2_score(y_test, y_pred_test)

print(f'Test MSE: {test_mse}')
print(f'Test R^2: {test_r2}')

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_test, alpha=0.5)
plt.plot([0, 100], [0, 100], '--r')
plt.xlabel('Actual Song Popularity')
plt.ylabel('Predicted Song Popularity')
plt.title('Testing Set: Actual vs Predicted')
plt.show()
```
